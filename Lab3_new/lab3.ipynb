Hãy khảo sát tỉ lệ dữ liệu thiếu trên biến huyết áp. Dữ liệu bị thiếu thì hãy xử lý bằng 
phương pháp sau

 Thay thế bằng giá trị trung bình liền trước và liền sau của người đó. Nếu không 
được thì dùng 2)

 Thay thế bằng giá trị trung bình 2 giá liền trước của người đó. Nếu không được 
thì dùng 3)

 Thay thế bằng giá trị trung bình 2 giá liền sau của người đó. Nếu không được thì 
dùng 4)

 Trung bình của các giá trị huyết áp của người đó. Nếu không được thì dùng 5).

 Trung bình của các giá trị huyết áp của nhóm giới tính. Nếu không được thì dùng 
6)

 Trung bình của các giá trị dữ liệu. Nếu không được thì thay bằng mức ổn định 
trong y học.

```python
import pandas as pd

# Đọc file CSV
try:
    df = pd.read_csv('patient_heart_rate.csv')
    print("Các cột trong dữ liệu:", df.columns)
    
    # Chuyển đổi cột blood_pressure sang dạng số, các giá trị không thể chuyển đổi sẽ trở thành NaN
    df['blood_pressure'] = pd.to_numeric(df['blood_pressure'], errors='coerce')
    print(df.head())
except Exception as e:
    print(f"Lỗi khi đọc file CSV: {e}")


```

```python
try:
    bp_column = 'blood_pressure'  # Sử dụng tên cột đúng
    missing_bp = df[bp_column].isnull().mean()
    print(f'Tỉ lệ dữ liệu thiếu trên biến huyết áp: {missing_bp:.2%}')
except KeyError as e:
    print(f"Lỗi: Cột '{bp_column}' không tồn tại trong dữ liệu: {e}")
except Exception as e:
    print(f"Lỗi khi khảo sát dữ liệu thiếu: {e}")
```

```python
def fill_missing_bp(df, bp_column):
    for idx in df[df[bp_column].isnull()].index:
        # Thay thế bằng giá trị trung bình liền trước và liền sau
        if idx > 0 and idx < len(df) - 1:
            prev_bp = df.at[idx - 1, bp_column]
            next_bp = df.at[idx + 1, bp_column]
            if not pd.isnull(prev_bp) and not pd.isnull(next_bp):
                df.at[idx, bp_column] = (prev_bp + next_bp) / 2
                continue
        
        # Thay thế bằng giá trị trung bình 2 giá liền trước
        if idx > 1:
            prev_bp_1 = df.at[idx - 1, bp_column]
            prev_bp_2 = df.at[idx - 2, bp_column]
            if not pd.isnull(prev_bp_1) and not pd.isnull(prev_bp_2):
                df.at[idx, bp_column] = (prev_bp_1 + prev_bp_2) / 2
                continue
        
        # Thay thế bằng giá trị trung bình 2 giá liền sau
        if idx < len(df) - 2:
            next_bp_1 = df.at[idx + 1, bp_column]
            next_bp_2 = df.at[idx + 2, bp_column]
            if not pd.isnull(next_bp_1) and not pd.isnull(next_bp_2):
                df.at[idx, bp_column] = (next_bp_1 + next_bp_2) / 2
                continue
        
        # Trung bình của các giá trị huyết áp của người đó
        patient_id = df.at[idx, 'patient_id']
        patient_bps = df[df['patient_id'] == patient_id][bp_column]
        patient_mean_bp = patient_bps.mean()
        if not pd.isnull(patient_mean_bp):
            df.at[idx, bp_column] = patient_mean_bp
            continue
        
        # Trung bình của các giá trị huyết áp của nhóm giới tính
        gender = df.at[idx, 'gender']
        gender_bps = df[df['gender'] == gender][bp_column]
        gender_mean_bp = gender_bps.mean()
        if not pd.isnull(gender_mean_bp):
            df.at[idx, bp_column] = gender_mean_bp
            continue
        
        # Trung bình của các giá trị dữ liệu
        overall_mean_bp = df[bp_column].mean()
        if not pd.isnull(overall_mean_bp):
            df.at[idx, bp_column] = overall_mean_bp
        else:
            # Thay bằng mức ổn định trong y học (ví dụ: 120)
            df.at[idx, bp_column] = 120

    return df

# Áp dụng hàm xử lý dữ liệu thiếu
try:
    df = fill_missing_bp(df, bp_column)
    print("Dữ liệu sau khi xử lý thiếu:")
    print(df.head())
except Exception as e:
    print(f"Lỗi khi xử lý dữ liệu thiếu: {e}")
```

```python
# Lưu kết quả sau khi xử lý vào file mới
try:
    df.to_csv('patient_heart_rate_filled.csv', index=False)
    print("Dữ liệu đã được lưu thành công.")
except Exception as e:
    print(f"Lỗi khi lưu dữ liệu: {e}")
```
