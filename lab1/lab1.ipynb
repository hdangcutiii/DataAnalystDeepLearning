```python
#import thư viện
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dulieuxettuyendaihoc.csv', index_col ="STT")

df.head(10)
```

```python
#4
df["DT"].fillna(0, inplace = False)
```

```python
#5
df["T1"].fillna(df ["T1"]. mean(), inplace = False)
```

```python
#6
is_float64 = df.dtypes == float

df_column_list = df.columns[is_float64].tolist()

df_column_list
```

```python
#7
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
df['TBM2'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
df['TBM3'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
print(df)
```

```python
#7
for i in ["1","2","6"]:
  df ["TBM"+i] = (df ["T"+i]*2 + df ["L"+i] + df ["H"+i] + df ["S"+i] + df ["V"+i]*2 + df ["X"+i] + df ["D"+i] + df ["N"+i]) / 10
  df.rename(columns={"TBM6":"TBM3"}, inplace=False)
```

```python
  df.head()
  #7
```

```python
#8
def classify_score(score):
    if score < 5.0:
        return 'Y'
    elif 5.0 <= score < 6.5:
        return 'TB'
    elif 6.5 <= score < 8.0:
        return 'K'
    elif 8.0 <= score < 9.0:
        return 'G'
    else:
        return 'XS'

df['XL1'] = df['TBM1'].apply(classify_score)
df['XL2'] = df['TBM2'].apply(classify_score)
df['XL3'] = df['TBM3'].apply(classify_score)
print(df)
```

```python
#9
df['US_TBM1'] = df['TBM1'] * (4/10)
df['US_TBM2'] = df['TBM2'] * (4/10)
df['US_TBM3'] = df['TBM3'] * (4/10)
print(df)
```

```python
#10
def determine_admission_result(row):
    if row['KT'] in ['A', 'A1']:
        if ((row['DH1']*2 + row['DH2'] + row['DH3']) / 4) >= 5.0:
            return 1
        else:
            return 0
    elif row['KT'] == 'B':
        if ((row['DH1'] + row['DH2']*2 + row['DH3']) / 4) >= 5.0:
            return 1
        else:
            return 0
    else:
        if ((row['DH1'] + row['DH2'] + row['DH3']) / 3) >= 5.0:
            return 1
        else:
            return 0


df['KQXT'] = df.apply(determine_admission_result, axis=1)


print(df)
```

```python
#11
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
```
